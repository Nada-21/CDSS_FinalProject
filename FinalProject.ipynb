```python
import os
import numpy as np
import pandas as pd
from scipy import signal

for i in range (1,9):
    dir_path = r"C:\Users\power\Desktop\CDSS_FinalProject\Fingers\S"+str(i)
    print('\n....................................................... Subject_0'+str(i)+" .................................................................")

    for file_name in os.listdir(dir_path):
        if file_name.endswith(".csv"):
            file_path = os.path.join(dir_path, file_name)
            df = pd.read_csv(file_path)
            result1=[]
            result2=[]
            # Load EMG signal from CSV file
            for i, col_name in enumerate(df.columns):
                emg_signal = df.iloc[:, i].values
                col_name_str = str(col_name)
                emg_signal = df[col_name_str].values
                # Compute time domain features
                mean = np.mean(emg_signal)
                median = np.median(emg_signal)
                std = np.std(emg_signal)
                var = np.var(emg_signal)
                rms = np.sqrt(np.mean(np.square(emg_signal)))
                skewness = np.mean((emg_signal - mean)**3) / np.power(std, 3)
                kurtosis = np.mean((emg_signal - mean)**4) / np.power(std, 4)
                maximum = np.max(emg_signal)
                minimum = np.min(emg_signal)
                range1 = maximum - minimum
                crest_factor = maximum / rms
                shape_factor = rms / np.mean(np.abs(emg_signal))
                var_abs = np.var(np.abs(emg_signal))
                threshold = np.mean(emg_signal)
                willison_amp = np.sum(np.abs(np.diff(emg_signal) > threshold))

                # Compute frequency domain features
                fs = 1000                     # Sampling frequency in Hz
                f, psd = signal.welch(emg_signal, fs=fs, nperseg=1024)
                spectral_centroid = np.sum(f * psd) / np.sum(psd)
                dominant_freq = f[np.argmax(psd)]
                mean_freq = np.sum(f * psd) / np.sum(psd)
                cumulative_sum = np.cumsum(psd)
                median_freq = f[np.argmin(np.abs(cumulative_sum - cumulative_sum[-1]/2))]
                bandwidth_90 = np.sum(psd[f <= 0.9*dominant_freq]) / np.sum(psd)
                entropy = -np.sum(psd * np.log2(psd))
                zero_crossing_rate = np.sum(np.abs(np.diff(np.sign(emg_signal)))) / (2 * len(emg_signal) / fs)
                features_dict1={'Electrode': i+1, 'Mean': mean, 'Median': median, 'Std': std, 'Var': var, 'RMS': rms,
                                'skewness':skewness,'Maximum':maximum,'Minimum':minimum,'Range':range1,'crest factor':crest_factor,
                                'shape factor': shape_factor,'Variance of absolute Vales':var_abs,'Willison Amplitude':willison_amp}
                result1.append(features_dict1)
                features_dict2={'Electrode': i+1,'spectral centroid ':spectral_centroid , 'Dominant Frequency':dominant_freq, 'Mean Frequency':mean_freq, 'Median Frequency':median_freq,
                                'Bandwidth (90%)':bandwidth_90,'Entropy':entropy,'Zero Crossing Rate': zero_crossing_rate}
                
                result2.append(features_dict2)
            
            print("Results for file: ", file_name)
            print('Time Domain Features: ')
            results_df1 = pd.DataFrame(result1)
            display(results_df1)
            print('\nFrequency Domain Features: ')
            results_df2 = pd.DataFrame(result2)
            display(results_df2)
```
